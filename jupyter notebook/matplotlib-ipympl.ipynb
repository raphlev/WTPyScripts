```python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));

# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
# Simple Python program to understand the Scatter Plot of the matplotlib with           # Jupyter  Notebook  
# Here, we are importing the pyplot module as pt from matplotlib library   
from matplotlib import pyplot as pt  
# here, we are giving the x-axis values to be plot on the graph  
x = [ 4, 2, 5, 3, 7 ]  
# here, we are giving the Y-axis values to be plot on the graph  
y = [ 7, 5, 9, 6, 3 ]  
# here, we are declaring the function to plot the Scatter on the graph  
pt.scatter(x, y)  
# here, we are adding the title to our graph  
pt.title("Matplotlib Graph")  
# here, we are labelling the axis  
pt.xlabel("Capacity")  
pt.ylabel("Liters")  
# here, we are declaring the function to show the plot on the graph  
pt.show()  
```
